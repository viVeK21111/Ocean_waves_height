```python
# as we see here, location doesn't much contribute to the height of the wave, so we will remove it
df_n=df_n.drop('LOCATION',axis=1)
```

```python
df_n
```

```python
# performing label encoding on the categorical data column
from sklearn.preprocessing import LabelEncoder
en = LabelEncoder()
df_n['COUNTRY']=en.fit_transform(df_n['COUNTRY'])
df_n['STATE/PROVINCE']=en.fit_transform(df_n['STATE/PROVINCE'])
```

```python
df_n.head()
```

```python
# Drop all columns with 'Unnamed:' in their name
df_n = df_n.loc[:, ~df_n.columns.str.contains('^Unnamed')]
```

```python
# checking outliers in the columns
plt.figure(figsize=(10,6))
for i in list(df_n.columns):
    sns.boxplot(df_n[i])
    plt.show()
```

```python
# we can use iqr method to remove the outliers
def Outlier_Treatment(df1,variable):
    percentile25=df1[variable].quantile(0.25)  #q1
    percentile75=df1[variable].quantile(0.75)  #q3
    iqr=percentile75-percentile25
    upper_limit=percentile75+1.5*iqr
    lower_limit=percentile25-1.5*iqr
    df_new = df1[ (df1[variable]>=lower_limit) & (df1[variable]<=upper_limit) ] 
    plt.figure(figsize=(12,6))
    sns.boxplot(df_new[variable])
    plt.show()
    return df_new

df_new = Outlier_Treatment(df_n,'YEAR')
df_new = Outlier_Treatment(df_n,'REGION_CODE')
df_new = Outlier_Treatment(df_new,'COUNTRY')
df_new = Outlier_Treatment(df_n,'STATE/PROVINCE')
df_new = Outlier_Treatment(df_new,'TRAVEL_TIME_MINUTES')
df_new = Outlier_Treatment(df_new,'DISTANCE_FROM_SOURCE')
df_new = Outlier_Treatment(df_new,'LATITUDE')
df_new = Outlier_Treatment(df_new,'MAXIMUM_HEIGHT')
```

```python
# NOTE: The outliers may not be removed fully, because the new outliers may form again, so execute the outlier funnction i.e iteratively removing outliers to get new data
# with no outliers
# plot the boxplot again to check there are no points outside the whiskers
```

```python
# execute multiple times to remove the outliers, until there are no points outside the whisker
df_new = Outlier_Treatment(df_new,'YEAR')
df_new = Outlier_Treatment(df_n,'REGION_CODE')
df_new = Outlier_Treatment(df_n,'LATITUDE')
df_new = Outlier_Treatment(df_new,'DISTANCE_FROM_SOURCE')
df_new = Outlier_Treatment(df_new,'MAXIMUM_HEIGHT')
```

```python
# selecting attributes based on univariate selection

from sklearn.feature_selection import SelectKBest,f_regression

x = df_new.iloc[:,:-2]
y= df_new.iloc[:,-2]  # continous numerical data
x['HORIZONTAL_INUNDATION']=df_new['HORIZONTAL_INUNDATION']
```

```python
from numpy import set_printoptions
X = x.values
Y = y.values
test = SelectKBest(score_func=f_regression, k=4) # chi2 is used for categorical value
fit = test.fit(X, Y)
set_printoptions(precision=3)
for i in fit.scores_:
    print(i)
```
